```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("population.csv")
data
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
countries=["Africa Eastern and Southern","Canada","Brazil","Germany","Zambia"]
filtered_data=data[data["Country Name"].isin(countries)]
plt.bar(filtered_data["Country Name"],filtered_data["2021"])
plt.xlabel("Country Name")
plt.ylabel("Population in 2021")
plt.title("Selected Countries in 2021")
plt.xticks(rotation=45)
plt.show()
```

```python
total_population=data[data["Indicator Code"]=="SP.POP.TOTL"]
```

```python
total_population_sort=total_population.sort_values(by="2021",ascending=False)
total_population_sort
```

```python
total_highest_population=total_population_sort.head(5)
total_highest_population
```

```python
plot = sns.barplot(x="2021", y="Country Name", data=total_highest_population)
plt.title("Top 10 countries of 2021 in Total Population")
plt.xlabel("Population")
plt.show()
```

```python
plot=sns.lineplot(x="2002",y="Country Name",data=total_highest_population)
plt.title("Top 10 countries of 2002 in Total Population")
plt.xlabel("Population")
plt.ylabel("Country")
```

```python
data["Average Growth Rate"]=data.loc[:,"2012":"2022"].mean(axis=1)
```

```python
plt.figure(figsize=(12, 8))
plt.barh(data["Country Name"].head(10), data["Average Growth Rate"].head(10))
plt.xlabel("Average Growth Rate")
plt.ylabel("Country")

plt.show()
```
